```python
aUrl = 'http://query1.finance.yahoo.com/v7/finance/download/%5EGSPC?period1=-1325635200&period2=1588550400&interval=1d&events=history'
import pandas as pd
df = pd.read_csv(aUrl)
df
```

```python
import matplotlib.pyplot as pl
ys = df['Close'].values
pl.plot(ys);pl.grid(True)
```

```python
ys = ys[-5000:]
pl.plot(ys);pl.grid(True)
```

```python
import thinkdsp as td
w= td.Wave(ys, framerate=8000)
w.plot()
```

```python
w.make_audio()
```

```python
dw= w.diff()
dw.plot()
dw.make_audio()
```
